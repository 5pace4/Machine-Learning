## Load Dataset from the following path

/kaggle/input/credit-card-customer-churn-prediction/Churn_Modelling.csv


```python
import numpy as np
import pandas as pd
path = r'/kaggle/input/credit-card-customer-churn-prediction/Churn_Modelling.csv'
df = pd.read_csv(path)
```

## Check the null value

```python
df.info()
```

## Statistics on the dataset 

```python
df.describe()
```

## Drop the unnecessary column

```python
df.drop(columns = ['RowNumber', 'CustomerId', 'Surname'], inplace = True)
```

```python
df.head()
```

## care about categorial data

```python
df = pd.get_dummies(df, columns = ['Geography', 'Gender'], drop_first = True)
```

```python
df.head()
```

## separate input and output features

```python
X = df.drop(columns = ['Exited'])
y = df['Exited']
```

```python
X
```

```python
y
```

## scaling input data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

## Split train-test data 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2, random_state = 42)
```

```python
X_train.shape
```

## Build an ANN(sequential) model

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(64, activation = 'relu', input_shape = (X_train.shape[1], )))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(16, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

### compile and fit the model

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model.fit(X_train, y_train, epochs = 10, validation_split = 0.2)
```

## Evaluate the model and get the accuray of the model

```python
model.evaluate(X_test, y_test)
```
